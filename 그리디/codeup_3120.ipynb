# codeup_3120 : 리모컨

컴퓨터실에서 수업 중인 정보 선생님은 냉난방기의 온도를 조절하려고 한다.

냉난방기가 멀리 있어서 리모컨으로 조작하려고 하는데, 리모컨의 온도 조절 버튼은 다음과 같다.

1) 온도를 1도 올리는 버튼

2) 온도를 1도 내리는 버튼

3) 온도를 5도 올리는 버튼

4) 온도를 5도 내리는 버튼

5) 온도를 10도 올리는 버튼

6) 온도를 10도 내리는 버튼

이와 같이 총 6개의 버튼으로 목표 온도를 조절해야 한다.

현재 설정 온도와 변경하고자하는 목표 온도가 주어지면 이 버튼들을 이용하여 목표 온도로 변경하고자 한다.

이 때 **버튼 누름의 최소 횟수** 를 구하시오.

예를 들어, 7도에서 34도로 변경하는 경우,

7 -> 17 -> 27 -> 32 -> 33 -> 34

이렇게 총 5번 누르면 된다.


## 내 풀이

### 생각 1
답안 예시에 목표온도가 현재온도보다 높길래 그 경우와 (목표온도가 현재온도보다 낮은 경우 생각 못함) 현재온도가 목표온도보다 무조건 작아야한다고 생각해서  
목표온도에서 10씩 빼고, 10 못 빼면 5 빼고, 5 못 빼면 1빼는 식으로 했음

### 생각 2
생각1처럼 푼 후 채점해봤는데 현재온도:22, 목표온도:3 이렇게 나오고 모범 답이 3이 나오길래(난 6이 나왔음) 생각해보니 3 -> 13 -> 23 -> 22 이렇게 목표 온도보다 커져도 되는 거였음.  
그래서 현재온도가 목표온도와 10도 미만 차이날 때 규칙을 찾아 봄.


|현재온도|차이|목표온도|리모컨 클릭|
|---|---|---|---|
|13|1|14|+1|
|13|2|15|+1 +1|
|13|3|16|+1 +1 +1 or +5 -1 -1|
|13|4|17|+5 -1|
|13|5|18|+5|
|13|6|19|+5 +1|
|13|7|20|+5 +1 +1|
|13|8|21|+10 -1|
|13|9|22|+10 -1 -1|

- 차이가 8이상 날 경우 10을 더해준 후 1을 뺀다
- 차이가 3이상(7이하) 날 경우 5를 더해준 후 1을 더하거나 뺀다  --> 차이가 3인 경우 두 가지로 계산 방법이 있지만 나는 통일성을 주기 위해 5를 더해준 후 1을 두번 빼는 것으로 계산함
- 차이가 1이상(2이하) 날 경우 1을 더하거나 뺌.

** 최종적으로 현재온도와 목표온도가 -2~+2 차이가 나서 여기만 특별히 elif문을 두 개 써줌



```python
# 입력값 : 현재온도, 목표온도

temp = list(map(int, input().split()))
a = max(temp)
b = min(temp)
click = 0

while True:
    while a >= (b+10):
        b += 10
        click +=1

    if (a-b)>=8:
        b += 10
        click +=1
    elif (a-b)>=3:
        b += 5
        click +=1
    elif (a-b)>=1:
        b +=1
        click +=1
    elif (b-a)>=1:
        b -= 1
        click +=1 
    elif a ==b:
        break
            
print(click)
```
